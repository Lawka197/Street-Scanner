```python
import pandas as pd 
import os 

filepath = '/Users/olivialawinski/IKT/analytics/Semesterabgabe/raw_data/combined_dataset_with_ab.csv'
#messdaten = pd.read_csv(filepath, sep= ',') 
#messdaten = pd.read_csv(filepath, sep=',', error_bad_lines=False)
messdaten = pd.read_csv(filepath, sep=';', on_bad_lines='skip')
```

```python
messdaten.head(10)
```

```python
messdaten['SENSOR_TYPE'] = pd.to_numeric(messdaten['SENSOR_TYPE'],errors = 'coerce')

# Filtern nach 20 
filtered_data = messdaten[messdaten['SENSOR_TYPE']== 20]

#Anzeigen
filtered_data.head(10)
```

```python
# aus der Spalte MEASUREMENT_VALUE den z-Wert extrahieren 
def extract_z_value(item):
    if isinstance(item,str) and "," in item:
        item = item.split(",")
        if len(item) == 3:
            return item[2]
    else:
        return item

# neue Spalte erstellen und die z-Werte dort speichern 
filtered_data.loc[:, "Z_VALUE"] = filtered_data["MEASUREMENT_VALUE"].map(extract_z_value)
filtered_data.sample(10)
```

```python
filtered_data = filtered_data.drop(columns = ['DEVICE_NAME','GPS_LAT','GPS_LONG'])
filtered_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

#Daten für Machine Learning vorbereiten 
features = filtered_data[['Z_VALUE']].dropna().astype(float)
labels = filtered_data['SOURCE_FOLDER']

# Trainings- und Testdaten aufteilen
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Modell initialisieren und trainieren
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Vorhersagen für Testdaten treffen
y_pred = clf.predict(X_test)

# Modell evaluieren
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

Die Confusion Matrix gibt eine Übersicht über die Vorhersagegenauigkeit für jede Klasse.
Zum Beispiel Die erste Zeile (Glatt): 40.188 Proben wurden korrekt als Glatt klassifiziert, aber 3.362 Proben, die tatsächlich Glatt sind, wurden als Uneben vorhergesagt.
Die zweite Zeile (Glatt/alte): 21.915 Proben, die tatsächlich Glatt/alte sind, wurden als Glatt vorhergesagt, aber keine wurde korrekt als Glatt/alte klassifiziert.
Der Classification Report:
Precision: Die Precision ist in den meisten Klassen sehr niedrig (oft 0.0), was bedeutet, dass das Modell viele falsche Positive hat.
Recall: Der Recall ist für Glatt relativ hoch (0.87), aber für die meisten anderen Klassen sehr niedrig, was bedeutet, dass das Modell viele tatsächliche Positive übersieht.
F1-Score: Der F1-Score ist ebenfalls sehr niedrig, was auf eine schlechte Balance zwischen Precision und Recall hinweist.
Accuracy: Die Gesamtgenauigkeit des Modells liegt bei 23%, was darauf hinweist, dass das Modell nur in 23% der Fälle korrekt vorhersagt.

Macro avg: Der ungewichtete Durchschnitt über alle Klassen. Hier sind Precision, Recall und F1-Score sehr niedrig (0.07, 0.11, 0.07), was auf eine schlechte Leistung hinweist.
Weighted avg: Der gewichtete Durchschnitt, der die Häufigkeit jeder Klasse berücksichtigt. Auch hier sind die Werte niedrig, was auf eine insgesamt schlechte Modellleistung hinweist.

```python
# Oversampling =  SMOTE (Synthetic Minority Over-sampling Technique)
# Das Ergebnis zeigt, dass SMOTE erfolgreich die Klassen im Datensatz ausgeglichen hat. 
# Jede Klasse hat nun 231.240 Instanzen, was ein ausgeglichenes Trainingsset sicherstellt. 
# Das ist ein guter Ausgangspunkt, um ein robustes Klassifikationsmodell zu trainieren.
```

```python
from imblearn.over_sampling import SMOTE

# SMOTE anwenden
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(features, labels)

# Überprüfen der neuen Verteilung der Klassen
print(y_resampled.value_counts())
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
# Trainings- und Testdaten aufteilen
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Modell initialisieren und trainieren
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Vorhersagen für Testdaten treffen
y_pred = clf.predict(X_test)

# Modell evaluieren
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Grid Search für Hyperparameter-Tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Beste Hyperparameter
print("Best Parameters:", grid_search.best_params_)

# Modell mit den besten Hyperparametern trainieren
best_clf = grid_search.best_estimator_
y_pred_best = best_clf.predict(X_test)

# Modell evaluieren
print("Confusion Matrix with Best Parameters:")
print(confusion_matrix(y_test, y_pred_best))
print("\nClassification Report with Best Parameters:")
print(classification_report(y_test, y_pred_best))
print("\nAccuracy Score with Best Parameters:")
print(accuracy_score(y_test, y_pred_best))
```

```python
# Beste Hyperparameter abrufen
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Modell mit den besten Hyperparametern trainieren
best_model = grid_search.best_estimator_
best_model.fit(X_train, y_train)

# Vorhersagen für Testdaten treffen
y_pred_best = best_model.predict(X_test)

# Modell evaluieren
print("Confusion Matrix with Best Parameters:")
print(confusion_matrix(y_test, y_pred_best))
print("\nClassification Report with Best Parameters:")
print(classification_report(y_test, y_pred_best))
print("\nAccuracy Score with Best Parameters:")
print(accuracy_score(y_test, y_pred_best))
```
